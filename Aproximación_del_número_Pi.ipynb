```python
import numpy as np
import random as rnd
import math
import matplotlib.pyplot as plt
```

```python
#Este es el experimento
x = np.linspace(0,1,100)
u_1 = rnd.random()
u_2 = rnd.random()
plt.plot(x, np.sqrt(1-x**2))
plt.scatter(u_1,u_2,color="red")
plt.show()
```

```python
def monte_carlo_pi(aprox_iteraciones):
    dentro_circulo = 0
    total_puntos = aprox_iteraciones

    for i in range(total_puntos):
        x_1 = np.linspace(0,1,100)
        u_1 = rnd.random()
        u_2 = rnd.random()
        #u_1 = rnd.uniform(0, 1)
        #u_2 = rnd.uniform(0, 1)

        if u_1**2 + u_2**2 <= 1:
            dentro_circulo += 1

    pi_approx = 4 * (dentro_circulo / total_puntos)
    plt.plot(x_1, np.sqrt(1-x_1**2))
    plt.scatter(u_1,u_2,color="red")
    plt.show()
    return pi_approx
```

```python
monte_carlo_pi(100)
```

```python
def experimento(iteraciones):
    i = 0
    dentro_circulo1=0
    Lista = []
    pi_approx = monte_carlo_pi(iteraciones)
    while i <= iteraciones:
        dentro_circulo1 = dentro_circulo1 + pi_approx
        i += 1
        probabilidad = dentro_circulo1/i
        Lista.append(probabilidad)
    return Lista
```

```python
exp = experimento(500000)
```

```python
np.mean(exp)
```

```python
plt.plot(exp)
plt.xlabel('Iteraciones')
plt.ylabel('Valor aproximado')
plt.title('Aproximación de Pi')
plt.show()
```
