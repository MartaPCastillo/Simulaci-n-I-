###Ejemplo 1

Puede ser simulada por el método de inversión mediante el siguiente algoritmo:

1. Generar $U \sim U(0, 1)$
2. Devolver:
 $X = 3 \cdot \sqrt[3]{U}$

```python
import random                                   # importamos el módulo random
import numpy as np                              # importamos numpy con un alias (np)
import matplotlib.pyplot as plt                 # ídem
#from random import expovariate, gauss
from statistics import mean, median, variance, stdev      # exportamos funciones especificas de statistic
```

```python
def ejem1(n):
    lista = []
    for i in range(n):
      u=random.random()
      X = 3*((u)**(1/3))
      lista.append(X)
    return lista
```

```python
G=ejem1(10000)
print(G)
```

```python
mu=np.mean(G)
print(mu)
```

```python
sigma2=np.var(G)
print(sigma2)
```

```python
plt.hist(G, bins=50)
```

```python
plt.plot(G)
```

```python
def cdf(muestra):
  m=len(muestra)
  x=np.sort(muestra)
  y=np.arange(0,m)/m
  return x,y
```

```python
X,Y = cdf(G)
```

```python
plt.step(X,Y)
```

###Ejemplo 2

###Ejemplo 3

Puede ser simulada por el método de composición mediante el siguiente algoritmo:

1. Generar $U_1, U_2 \sim U(0, 1)$
2. Si $U_1 \in (0,1/2)$:
      $X = \dfrac{\ln{(2 \cdot U_2)}}{2}$
3. Otro caso:
      $X = -\dfrac{\ln{(1 - 2 \cdot U_2)}}{2}$

```python
import random                                   # importamos el módulo random
import numpy as np                              # importamos numpy con un alias (np)
import matplotlib.pyplot as plt                 # ídem
#from random import expovariate, gauss
from statistics import mean, median, variance, stdev      # exportamos funciones especificas de statistic
```

```python
def ejem3(n):
    lista = []
    for i in range(n):
      R_1=random.random()
      R_2 = random.random()
      if 0 <= R_1 <= 1/2:
         X = (np.log(2*R_2))/2
      else:
         X = -(np.log(1 - (2*R_2)))/2
      lista.append(X)
    return lista
```

```python
H=ejem2(10000)
print(H)
```

```python
mu=np.mean(H)
print(mu)
```

```python
sigma2=np.var(H)
print(sigma2)
```

```python
plt.hist(H, bins=50)
```

```python
plt.plot(H)
```

```python
def cdf(muestra):
  m=len(muestra)
  x=np.sort(muestra)
  y=np.arange(0,m)/m
  return x,y
```

```python
X,Y = cdf(H)
```

```python
plt.step(X,Y)
```
