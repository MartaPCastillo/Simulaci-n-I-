###Ejemplo 1

Puede ser simulada por el método de inversión mediante el siguiente algoritmo:

1. Generar $U \sim U(0, 1)$
2. Devolver:
 $X = 3 \cdot \sqrt[3]{U}$

```python
import random                                   # importamos el módulo random
import numpy as np                              # importamos numpy con un alias (np)
import matplotlib.pyplot as plt                 # ídem
#from random import expovariate, gauss
from statistics import mean, median, variance, stdev      # exportamos funciones especificas de statistic
```

```python
def ejem1(n):
    lista = []
    for i in range(n):
      u=random.random()
      X = 3*((u)**(1/3))
      lista.append(X)
    return lista
```

```python
G=ejem1(10000)
print(G)
```

```python
mu=np.mean(G)
print(mu)
```

```python
sigma2=np.var(G)
print(sigma2)
```

```python
plt.hist(G, bins=50)
```

```python
plt.plot(G)
```

```python
def cdf(muestra):
  m=len(muestra)
  x=np.sort(muestra)
  y=np.arange(0,m)/m
  return x,y
```

```python
X,Y = cdf(G)
```

```python
plt.step(X,Y)
```
