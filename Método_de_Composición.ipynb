### 6. Repita el experimento $n$ veces

Así dispondrá de $n$ observaciones sobre el comportamiento del sistema, lo que permitirá entender su funcionamiento así como evaluar el desempeño del mismo frente a los diversos escenarios establecidos.

El **método Monte Carlo** se basa en la **Ley Fuerte de los Grandes Números** y el **Teorema del Límite Central**.

```python
import random                                   # importamos el módulo random
import numpy as np                              # importamos numpy con un alias (np)
import matplotlib.pyplot as plt                 # ídem
#from random import expovariate, gauss
from statistics import mean, median, variance, stdev      # exportamos funciones especificas de statistic
```

```python
def met_comp(a,b,c,n):
    lista = []
    m = (b-a)/(c-a)
    for i in range(n):
      R_1=random.random()
      R_2 = random.random()
      if 0 <= R_1 <= m:
         X= a + (c-a)*(np.sqrt(R_2))
      else:
         X= c - (c-b)*(np.sqrt(R_2))
      lista.append(X)
    return lista
```

```python
G=met_comp(2,5,8,1000)
print(G)
```

```python
mu=np.mean(G)
print(mu)
```

```python
sigma2=np.var(G)
print(sigma2)
```

```python
plt.hist(G, bins=50)
```

```python
plt.plot(G)
```

```python
def cdf(muestra):
  m=len(muestra)
  x=np.sort(muestra)
  y=np.arange(0,m)/m
  return x,y
```

```python
X,Y = cdf(G)
```

```python
plt.step(X,Y)
```
